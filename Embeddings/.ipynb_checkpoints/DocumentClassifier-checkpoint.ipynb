```python
# specify the folder that contains directories with the same name as the classification label
folder = "./classify"
# specif the name of the file to save the training data (text plus labels)
training_data_filename = 'document_classifier_training.txt'
model_name = 'document_classifier.bin'
```

## Helper Functions

```python
# specify folder container directory structure for training
def get_folder_names(path):
    """
    Get folder names in a file directory
    """
    return [name for name in os.listdir(path)
            if os.path.isdir(os.path.join(path, name)) and name not in ['.ipynb_checkpoints']]

def get_files_recursive(dir_path):
    files = []
    for (dirpath, dirnames, filenames) in os.walk(dir_path):
        files.extend(filenames)
        for dirname in dirnames:
            dir = os.path.join(dirpath, dirname)
            files.extend(get_files_recursive(dir))
    return files

def get_text(filename, label):
    parsed = parser.from_file(filename, requestOptions={'headers': headers, 'timeout': 300})
    text = parsed["content"]
    if text is not None:
        clean_text = " ".join(text.split())
        return label +" "+clean_text
    return text

def save_list(list_to_save, file_name):
    with open(file_name, 'w') as file_handler:
        for item in list_to_save:
            file_handler.write("{}\n".format(item))
```

## get directories used for classification

```python
directories = get_folder_names(folder)
directories
```

## Get text from all files in directory as a single line of text and save to training.txt with classification category appended


```python
text_data = []
for directory in directories:
    files = get_files_recursive(folder+'/'+directory)
    for file in files:
        print("getting file content: ", file)
        category = "__label__"+directory
        text = get_text(folder+"/"+directory+"/"+file, category)
        text_data.append(text)
        
save_list(text_data, training_data_filename)
```

## Traing a fasttext model for classification

```python
# train the model
model = fasttext.train_supervised(input=training_data_filename, lr=0.33, epoch=100, wordNgrams=5, bucket=200000, dim=50, loss='hs')
```

## Save the model

```python
model.save_model(model_name)
```

```python
## Load the saved model
```

```python
model = fasttext.load_model(model_name)
```

# Use the model to classify new text

```python
result = model.predict("What type of file am I if this is the extractred text?")
predict_category = result[0][0].replace("__label__", "")
predict_category
```

# Starting from pretrained vectors often improves performance with limited data

```python
import fasttext.util
fasttext.util.download_model('en', if_exists='ignore')  # English
ft = fasttext.load_model('cc.en.300.bin')
```

### The pre-trained word vectors we distribute have dimension 300. If you need a smaller size, you can use our dimension reducer.

```python
ft = fasttext.load_model('cc.en.300.bin')
ft.get_dimension()
```

```python
fasttext.util.reduce_model(ft, 50)
ft.get_dimension()
ft.save_model('cc.en.50.bin')
```

### Need to convert the .bin file to a .vec file

```python
VECTORS_FILEPATH = 'cc.en.50.vec'
# original BIN model loading
f = fasttext.load_model('cc.en.50.bin')
lines=[]

# get all words from model
words = f.get_words()

with open(VECTORS_FILEPATH,'w') as file_out:
    
    # the first line must contain number of total words and vector dimension
    file_out.write(str(len(words)) + " " + str(f.get_dimension()) + "\n")

    # line by line, you append vectors to VEC file
    for w in words:
        v = f.get_word_vector(w)
        vstr = ""
        for vi in v:
            vstr += " " + str(vi)
        try:
            file_out.write(w + vstr+'\n')
        except:
            pass
```

### train a fasttext classifier starting from pretrained word vectors

```python
model = fasttext.train_supervised(input=training_data_filename, lr=.33, epoch=100,
                             wordNgrams=5, bucket=200000, dim=50, loss='hs',
                             pretrainedVectors=VECTORS_FILEPATH)
```
